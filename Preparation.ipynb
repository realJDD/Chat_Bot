# Train Classifier

```python
# BOW model
words = []
classes = []
documents = []
ignore_words = [".", ",", "?"]
```

```python
def clean_up_sentence(sentence):
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [word.lower() for word in sentence_words if word not in ignore_words]
    return sentence_words
```

```python
for intent in intents["intents"]:
    for sentence in intent["patterns"]:
        ws = clean_up_sentence(sentence)
        words.extend(ws)
        documents.append((ws, intent["tag"]))
        if intent["tag"] not in classes:
            classes.append(intent["tag"])

words = sorted(list(set(words)))
```

```python
# Create training data
training = []
output_empty = [0] * len(classes)

for doc in documents:
    bag = []
    pattern_words = doc[0]
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)
    
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    
    training.append([bag, output_row])

# Shuffle
random.shuffle(training)
```

```python
X = []
Y = []

for i in range(len(training)):
    X.append(training[i][0])
    Y.append(training[i][1])

X = np.array(X)
Y = np.array(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
# Build NN model
model = Sequential()
model.add(Dense(12, input_dim=len(words), activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(len(classes), activation="softmax"))
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, batch_size=4, epochs=50)
```

```python
# Save the model to a json file
model_json = model.to_json()
with open("model/model.json", "w") as json_file:
    json_file.write(model_json)
print("Saved model to disk...")
```

```python
# Serialize weights to HDF5
model.save_weights("weights/weights.h5")
print("Saved weights to disk...")
```

```python
def bow(sentence, words, show_details=False):
    sentence_words = clean_up_sentence(sentence)
    bag = [0] * len(words)
    for w in sentence_words:
        if w in words:
            bag[words.index(w)] = 1
            if show_details:
                print(f"Found in bag: {w}")
    
    return np.array(bag)
```

```python
# Quick test
print(bow("How does the trapezius work?", words, show_details=True))
```

```python
ERROR_THRESHOLD = 0.25

def classify(sentence):
    X_in = bow(sentence, words)
    X_in = X_in[np.newaxis, :]
    probs = model.predict(X_in)[0]
    results = [[i, p] for i, p in enumerate(probs) if p > ERROR_THRESHOLD]
    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append((classes[r[0]], r[1]))
    
    return return_list
```
