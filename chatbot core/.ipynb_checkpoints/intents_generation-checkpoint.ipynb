# Train Classifier

First try a word level model without stemming.

```python
# BOW model
words = []
classes = []
documents = []
ignore_words = [".", ",", "?"]

for intent in intents["intents"]:
    for sentence in intent["patterns"]:
        ws = nltk.word_tokenize(sentence)
        words.extend(ws)
        documents.append((ws, intent["tag"]))
        if intent["tag"] not in classes:
            classes.append(intent["tag"])            
```

```python
words
```

```python
classes
```

```python
documents
```

```python
len(documents)
```

```python
words = [word.lower() for word in words if word not in ignore_words]
words = sorted(list(set(words)))
```

```python
words
```

```python
len(words)
```

```python
# Create training data
training = []
output_empty = [0] * len(classes)

for doc in documents:
    bag = []
    pattern_words = doc[0]
    pattern_words = [word.lower() for word in pattern_words]
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)
    
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    
    training.append([bag, output_row])

# Shuffle
random.shuffle(training)
```

```python
X = []
Y = []

for i in range(len(training)):
    X.append(training[i][0])
    Y.append(training[i][1])

X = np.array(X)
Y = np.array(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
# Build NN model
model = Sequential()
model.add(Dense(12, input_dim=len(words), activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(len(classes), activation="softmax"))
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, batch_size=4, epochs=50)
```

```python
# Save the model to a json file
model_json = model.to_json()
with open("model/model.json", "w") as json_file:
    json_file.write(model_json)
print("Saved model to disk...")
```

```python
# Serialize weights to HDF5
model.save_weights("weights/weights.h5")
print("Saved weights to disk...")
```

```python
def clean_up_sentence(sentence):
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [word.lower() for word in sentence_words if word not in ignore_words]
    return sentence_words
```

```python
def bow(sentence, words, show_details=False):
    sentence_words = clean_up_sentence(sentence)
    bag = [0] * len(words)
    for w in sentence_words:
        if w in words:
            bag[words.index(w)] = 1
            if show_details:
                print(f"Found in bag: {w}")
    
    return np.array(bag)
```

```python
# Quick test
print(bow("How does the trapezius work?", words, show_details=True))
```

```python
ERROR_THRESHOLD = 0.25

def classify(sentence):
    X_in = bow(sentence, words)
    X_in = X_in[np.newaxis, :]
    probs = model.predict(X_in)[0]
    results = [[i, p] for i, p in enumerate(probs) if p > ERROR_THRESHOLD]
    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append((classes[r[0]], r[1]))
    
    return return_list
```

```python
def match_check(sentence, body_parts_muscles_list):
    classify_list = classify(sentence)
    sentence_class = classify_list[0][0]
    if sentence_class == "exercise" or sentence_class == "identification":
        keywords_list = []
        sentence_words = clean_up_sentence(sentence)
        for w in sentence_words:
            keywords_list.extend(get_close_matches(w, body_parts_muscles_list))
        keywords_list = list(set(keywords_list))
        print(f"key words list: {keywords_list}")
        if len(keywords_list) == 0:
            print("Seems like the key word in your sentence is not in my database! Please try some other key words!")
        else:
            perfect_match_list = []
            for w in keywords_list:
                if re.search(w, sentence):
                    perfect_match_list.append(w)
            if "biceps brachii" in perfect_match_list and "biceps" in perfect_match_list:
                perfect_match_list.remove("biceps")
            if "triceps brachii" in perfect_match_list and "triceps" in perfect_match_list:
                perfect_match_list.remove("triceps")
#             if "thighs" in perfect_match_list and "thigh" in perfect_match_list:
#                 perfect_match_list.remove("thigh")
            print(f"perfect match list: {perfect_match_list}")
            if len(perfect_match_list) == 0:
                print("Just to make sure, is the keyword in your question '" + keywords_list[0] + "' please?")
            elif len(perfect_match_list) == 1:
                if sentence_class == "exercise":
                    print(f"OK! Show the exercise for {perfect_match_list[0]} now!")
                if sentence_class == "identification":
                    print(f"OK! Show the location of {perfect_match_list[0]} on the SVG!")
            else:
                print("I see multiple keywords in your question, please pick one of them:")
                for w in perfect_match_list:
                    print(w)                    
```
